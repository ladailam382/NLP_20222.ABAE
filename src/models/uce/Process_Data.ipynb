```python
data=[]
for i in train:
    d=[x if x in wv.key_to_index else '<OOV>' for x in i]
    if len(d)==0:
      continue
    d_mean=np.mean(np.array([wv[x] for x in d]),0)
    s=[d_mean@x for x in seed_embedding]
    s_max=max(s)
    s_min=min(s)
    y=[1 if x ==s_max else 0 for x in s]
    s=[(x-s_min)/(s_max-s_min) for x in s]
    c=[1 if y[i]==1 else 1-x for i,x in enumerate(s)]
    data.append([d,y,c])
```
